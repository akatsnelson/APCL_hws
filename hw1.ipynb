## 3. Лемматизация Mystem vs Pymorphy (2 балла)

Лемматизируйте текст с помощью mystem и pymorphy. Найдите различия в лемматизации. Что по вашему работает лучше на приведенном тексте?

Важно: для пайморфи используйте токенизацию из mystem, чтобы исключить влияние токенизации на результат. Анализируйте только значимые различия, а не технические особенности (не сравнивайте скорость работы и удобность интерфейса).

В майстеме убедитесь, что используется дизамбигуация.

```python
from pymystem3 import Mystem
from pymorphy2 import MorphAnalyzer

mystem = Mystem()
mystem_analysis = mystem.analyze(text)
mystem_tokens = []
mystem_lemmas = []
for item in mystem_analysis:
    tok = item['text']
    if not tok.strip():
        continue
    mystem_tokens.append(tok)
    analyses = item.get('analysis')
    if analyses:
        mystem_lemmas.append(analyses[0]['lex'])
    else:
        mystem_lemmas.append(tok)

pym = MorphAnalyzer()

pymorphy_lemmas = []
for tok in mystem_tokens:
    p = pym.parse(tok)[0]
    pymorphy_lemmas.append(p.normal_form)

for tok, lem_m, lem_p in zip(mystem_tokens, mystem_lemmas, pymorphy_lemmas):
    if lem_m != lem_p:
        print(f"{tok!r}: mystem → {lem_m!r}, pymorphy → {lem_p!r}")
```

Mystem разбирает глаголы, сохраняет имена собственные и латиницу, не искажает смысл слов. У Pymorphy больше ошибок: он путает регистр, меняет значения и неправильно обрабатывает имена.

```python
import requests

url = "https://raw.githubusercontent.com/mannefedov/compling_nlp_hse_course/master/data/anna_karenina.txt"
text = requests.get(url).text

chunks = [chunk.strip() for chunk in text.split("\n\n") if len(chunk.strip()) > 100]
documents = chunks[:1000]
```

```python
def get_k_shingles(text, k=5):
    tokens = text.lower().split()
    return set([' '.join(tokens[i:i + k]) for i in range(len(tokens) - k + 1)])
```

```python
from datasketch import MinHash, MinHashLSH
from tqdm import tqdm

k = 3
num_perm = 128
bands = 16

minhashes = []
shingle_sets = []

for doc in tqdm(documents):
    shingles = get_k_shingles(doc, k)
    shingle_sets.append(shingles)
    m = MinHash(num_perm=num_perm)
    for s in shingles:
        m.update(s.encode('utf8'))
    minhashes.append(m)
```

```python
from datasketch import MinHashLSH

num_perm = 128
b = 32
r = 4

lsh = MinHashLSH(threshold=0.5, num_perm=num_perm)
```

```python
for i, m in enumerate(minhashes):
    lsh.insert(f"doc_{i}", m)
    
results = []
for i, m in enumerate(minhashes):
    candidates = lsh.query(m)
    for key in candidates:
        j = int(key.split("_")[1])
        if i < j:
            real_jaccard = len(shingle_sets[i] & shingle_sets[j]) / \
                            len(shingle_sets[i] | shingle_sets[j])
            if real_jaccard > 0:
                results.append((i, j, real_jaccard))

results = sorted(results, key=lambda x: -x[2])
for i, j, score in results[:10]:
    print(f"Pair ({i}, {j}) — Jaccard: {score:.3f}")
```
